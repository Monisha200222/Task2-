```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/customers-10.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['City'].astype(str), data['Country'].astype(str))
plt.title("Scatter plot")
plt.xlabel('City')
plt.ylabel('Country')
plt.show()
```

```python
import matplotlib.pyplot as plt
# Convert 'Company' column to string type, handling potential NaN values
plt.plot(data['Company'].astype(str))
plt.plot(data['City'].astype(str))
plt.ylabel('City')
plt.xlabel('Company')
plt.show()
#Line chart
```

```python
# prompt: bar chart plot

import matplotlib.pyplot as plt
plt.bar(data['Company'].astype(str), data['City'].astype(str))
plt.xlabel('Company')
plt.ylabel('City')
plt.title('Bar Chart')
plt.show()
```
